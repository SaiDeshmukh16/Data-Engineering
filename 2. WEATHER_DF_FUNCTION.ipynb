```python
def weather(c_list):

    df_list = []

    for city in c_list:
        API_key = "7c94895786f3d0292d6248ccfd58881e"
        
        url = f"http://api.openweathermap.org/data/2.5/forecast?q={city}&appid={API_key}&units=metric"
        
        response = requests.get(url)
        
        json = response.json()


        for i in json["list"]:
                
            weather_dict = {
                "rain": [],
                "city_name": [],
                "country": [],
                "main_temp": [],
                "feels_like_temp": [],
                "time": []
            }
            
            try:
                weather_dict["rain"].append(i["rain"]["3h"])
            except:
                weather_dict["rain"].append("None")
            weather_dict["city_name"].append(json['city']['name'])
            weather_dict["country"].append(json['city']['country'])
            weather_dict["main_temp"].append(i['main']['temp'])
            weather_dict["feels_like_temp"].append(i['main']['feels_like'])
            weather_dict["time"].append(i['dt_txt'])
            
            weather_dataframe = pd.DataFrame(weather_dict)
        
            weather_dataframe["time"] = pd.to_datetime(weather_dataframe["time"])
            
            
        
            df_list.append(weather_dataframe)
        
    return pd.concat(df_list)    

city_list = ['Berlin', 'Hamburg', 'London', 'Manchester', 'Barcelona']
weather_df = weather(city_list)
```

```python
weather_df = weather(city_list)
weather_df
```

```python
weather_df.loc[weather_df["city_name"] == "Berlin", "city_name"] = 1
weather_df.loc[weather_df["city_name"] == "Hamburg", "city_name"] = 2
weather_df.loc[weather_df["city_name"] == "London", "city_name"] = 3
weather_df.loc[weather_df["city_name"] == "Manchester", "city_name"] = 4
weather_df.loc[weather_df["city_name"] == "Barcelona", "city_name"] = 5

weather_df.rename(columns = {"city_name" : "cityid", "time" : "date_time"}, inplace = True)
weather_df
```

```python
def weather_b(c_list):
    df_list = []

    for city in c_list:
        
        API_key = "7c94895786f3d0292d6248ccfd58881e"
        url = f"http://api.openweathermap.org/data/2.5/forecast?q={city}&appid={API_key}&units=metric"
        response = requests.get(url)
        json = response.json()

        for item in json["list"]:
            weather_dict = {
                "city_name": [],
                "country": [],
                "rain": [],
                "temperature": [],
                "time_for_forecast": []
                }
            try:
                weather_dict["rain"].append(item["rain"]["3h"])
            except:
                weather_dict["rain"].append("0")
            weather_dict["city_name"].append(json["city"]["name"])
            weather_dict["country"].append(json["city"]["country"])
            weather_dict["temperature"].append(item["main"]["temp"])
            weather_dict["time_for_forecast"].append(item['dt_txt'])

            weather_df = pd.DataFrame(weather_dict)
            weather_df["time_for_forecast"] = pd.to_datetime(weather_df["time_for_forecast"])
            df_list.append(weather_df)

    return pd.concat(df_list)
```
